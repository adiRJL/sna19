```python

import networkx as nx
from node2vec import Node2Vec
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import numpy as np
import random
import math
from networkx import to_numpy_matrix
```

```python
EMBEDDING_FILENAME = './embeddings.emb'
EMBEDDING_MODEL_FILENAME = './embeddings.model'
```

```python
def gcn_layer(A_hat, D_hat, X, W):
    return relu(D_hat**-1 * A_hat * X * W)

def relu(X):
    return np.maximum(0,X)
```

```python
#Extracting dataframes

feat_names = pd.read_csv("facebook/0.featnames")
edges = pd.read_csv("facebook/0.edges")
feat = pd.read_csv("facebook/0.feat")
circles = pd.read_csv("facebook/0.circles")
```

```python
# building the graph

#print(edges)

#edgelist is needed for read_edgelist
edge_list = open("facebook/0.edges", "rb")
#print(list(edge_list))
G = nx.read_edgelist(edge_list)
plt.figure(figsize=(14,14))
nx.draw(G, node_size=80, node_color='yellow', with_labels=True)

        
```

```python
#Building dataframe for link prediction

# getting nodes and edges for dataframe from graph
print(len(G.nodes))
node_list = list(np.unique(G.nodes))
edge_list = list(G.edges)

print(len(edge_list))
#print(edge_list)

entries = []
for edge in edge_list:
    entries.append([edge[0], edge[1], 1])
#print(len(entries))


for i in range(0, len(node_list)):
    for j in range(i+1, len(node_list)):
            edge = (node_list[i], node_list[j])
            edge2 = (node_list[j], node_list[i])
            if edge not in edge_list and edge2 not in edge_list:
                entries.append([edge[0], edge[1], 0])

               
                


df = pd.DataFrame(entries, columns = ['source_node', 'dest_node', 'edge_exists'])      
print(df)
```

```python
#build train graph

train_graph = nx.Graph()
node_list = [int(node) for node in node_list]
train_graph.add_nodes_from(node_list)
train_edge_list = random.sample(edge_list, math.ceil(0.8*len(edge_list)))
train_edge_list = [(int(i),int(j)) for i,j in train_edge_list]
train_graph.add_edges_from(train_edge_list)

plt.figure(figsize=(14,14))
nx.draw(train_graph, node_size=80, node_color='yellow', with_labels=True)

# for i in range(len(X_train)):
#     print(X_train.iloc[i])
```

```python
# get node to vec embeddings
node2vec = Node2Vec(train_graph, dimensions=50, walk_length=5, num_walks=5, workers=4)  # Use temp_folder for big graphs

model = node2vec.fit(window=10, min_count=1, batch_words=4)  # Any keywords acceptable by gensim.Word2Vec can be passed, `diemnsions` and `workers` are automatically passed (from the Node2Vec constructor)

print(model.wv.most_similar('2'))
print(model)
```

```python
# Save embeddings for later use
model.wv.save_word2vec_format(EMBEDDING_FILENAME)

# Save model for later use
model.save(EMBEDDING_MODEL_FILENAME)

emb = pd.read_csv(EMBEDDING_FILENAME, "r")
print(emb)
```

```python

order = sorted(list(train_graph.nodes()))
A = to_numpy_matrix(train_graph, nodelist=order)
I = np.eye(train_graph.number_of_nodes())
A_hat = A + I
D_hat = np.array(np.sum(A_hat, axis=0))[0]
D_hat = np.matrix(np.diag(D_hat))

W_1 = np.random.normal(
    loc=0, scale=1, size=(train_graph.number_of_nodes(), 4))
W_2 = np.random.normal(
    loc=0, size=(W_1.shape[1], 2))

H_1 = gcn_layer(A_hat, D_hat, I, W_1)
H_2 = gcn_layer(A_hat, D_hat, H_1, W_2)
output = H_2

feature_representations = {
    node: np.array(output)[node] 
    for node in train_graph.nodes()}

print(feature_representations)
```
